```python
# Importar bibliotecas necessárias
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import json

# Ler os dados do arquivo JSON
with open('dados_final.json', 'r') as f:
    data = json.load(f)

# Extrair as altitudes e armazená-las em uma lista
altitudes = []
for entry in data:
    altitude = entry['gnss_data']['altitude']
    altitudes.append(altitude)

# Criar o histograma da altitude
plt.figure(figsize=(10, 6))
sns.histplot(altitudes, bins=20)
plt.title('Distribuição da Altitude')
plt.xlabel('Altitude')
plt.ylabel('Frequência')
plt.show()
```

```python
# Boxplot da Altitude
plt.figure(figsize=(10, 6))
sns.boxplot(y=altitudes)
plt.title('Boxplot da Altitude')
plt.ylabel('Altitude')
plt.show()
```

```python
# Extrair altitudes e tempos
altitudes = []
tempos = []
for entry in data:
    altitude = entry['gnss_data']['altitude']
    tempo = entry['gnss_data']['timestamp']
    altitudes.append(altitude)
    tempos.append(tempo)

# Scatter Plot da Altitude em relação ao Tempo
plt.figure(figsize=(10, 6))
plt.scatter(tempos, altitudes, alpha=0.5)
plt.title('Altitude ao longo do Tempo')
plt.xlabel('Tempo')
plt.ylabel('Altitude')
plt.show()
```

```python
# Contagem de Satélites por Entrada
num_satelites = [len(entry['gnss_data']['satellites']) for entry in data]
plt.figure(figsize=(10, 6))
sns.countplot(num_satelites)
plt.title('Contagem de Satélites por Entrada')
plt.xlabel('Número de Satélites')
plt.ylabel('Frequência')
plt.show()
```

```python
# Extrair latitudes, longitudes e altitudes
latitudes = []
longitudes = []
altitudes = []
for entry in data:
    latitude = entry['gnss_data']['latitudeDiferenca']
    longitude = entry['gnss_data']['longitudeDiferenca']
    altitude = entry['gnss_data']['altitude']
    latitudes.append(latitude)
    longitudes.append(longitude)
    altitudes.append(altitude)

# Scatter Plot da Altitude em relação à Latitude e Longitude
plt.figure(figsize=(10, 6))
plt.scatter(longitudes, latitudes, c=altitudes, cmap='viridis', alpha=0.5)
plt.colorbar(label='Altitude')
plt.title('Altitude em relação à Latitude e Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Inicializar listas de latitudes, longitudes e precisões
latitudes = []
longitudes = []
precisoes = []

# Iterar sobre cada entrada no JSON e extrair os dados relevantes
for entry in data:
    latitude = entry['gnss_data']['latitudeDiferenca']
    longitude = entry['gnss_data']['longitudeDiferenca']
    precisao = entry['gnss_data']['accuracy']
    latitudes.append(latitude)
    longitudes.append(longitude)
    precisoes.append(precisao)

# Scatter Plot da Precisão em relação à Latitude e Longitude
plt.figure(figsize=(10, 6))
plt.scatter(longitudes, latitudes, c=precisoes, cmap='inferno', alpha=0.5)
plt.colorbar(label='Precisão')
plt.title('Precisão em relação à Latitude e Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

```

```python
# Inicializar listas de latitudes, longitudes e velocidades
latitudes = []
longitudes = []
velocidades = []

# Iterar sobre cada entrada no JSON e extrair os dados relevantes
for entry in data:
    latitude = entry['gnss_data']['latitudeDiferenca']
    longitude = entry['gnss_data']['longitudeDiferenca']
    velocidade = entry['gnss_data']['speed']
    latitudes.append(latitude)
    longitudes.append(longitude)
    velocidades.append(velocidade)

# Scatter Plot da Velocidade em relação à Latitude e Longitude
plt.figure(figsize=(10, 6))
plt.scatter(longitudes, latitudes, c=velocidades, cmap='plasma', alpha=0.5)
plt.colorbar(label='Velocidade')
plt.title('Velocidade em relação à Latitude e Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

```

```python
# Inicializar listas de latitudes, longitudes e elevações dos satélites
latitudes = []
longitudes = []
elevacoes_satelites = []

# Iterar sobre cada entrada no JSON e extrair os dados relevantes
for entry in data:
    latitude = entry['gnss_data']['latitudeDiferenca']
    longitude = entry['gnss_data']['longitudeDiferenca']
    elevacoes = [satelite['elevationDegrees'] for satelite in entry['gnss_data']['satellites']]
    for elevacao in elevacoes:
        latitudes.append(latitude)
        longitudes.append(longitude)
        elevacoes_satelites.append(elevacao)

# Scatter Plot da Elevação dos Satélites em relação à Latitude e Longitude
plt.figure(figsize=(10, 6))
plt.scatter(longitudes, latitudes, c=elevacoes_satelites, cmap='viridis', alpha=0.5)
plt.colorbar(label='Elevação dos Satélites')
plt.title('Elevação dos Satélites em relação à Latitude e Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

```

```python
# Inicializar listas de latitudes, longitudes e precisões do sinal dos satélites
latitudes = []
longitudes = []
precisoes_sinal_satelites = []

# Iterar sobre cada entrada no JSON e extrair os dados relevantes
for entry in data:
    latitude = entry['gnss_data']['latitudeDiferenca']
    longitude = entry['gnss_data']['longitudeDiferenca']
    precisoes_sinal = [satelite['cn0DbHz'] for satelite in entry['gnss_data']['satellites']]
    
    # Adicionar os dados à lista principal apenas uma vez
    latitudes.extend([latitude] * len(precisoes_sinal))
    longitudes.extend([longitude] * len(precisoes_sinal))
    precisoes_sinal_satelites.extend(precisoes_sinal)

# Scatter Plot da Precisão do Sinal dos Satélites em relação à Latitude e Longitude
plt.figure(figsize=(10, 6))
plt.scatter(longitudes, latitudes, c=precisoes_sinal_satelites, cmap='inferno', alpha=0.5)
plt.colorbar(label='Precisão do Sinal dos Satélites')
plt.title('Precisão do Sinal dos Satélites em relação à Latitude e Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
